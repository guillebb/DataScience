2.Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df.columns
```

```python
datos_salario = df['Salary']
datos_salario
```

```python
mediana = datos_salario.median()
mediana
```

```python
datos_salario.values.flatten()
```

```python
datos = np.random.exponential()
plt.hist(datos_salario, bins= 5, align = 'mid', orientation = 'vertical', color = 'pink', ec = 'black')
plt.title('Salarios Mas Populares')
plt.show()
```

4.Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)

```python
df.dtypes
```

```python
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df = df.set_index('DateofHire')
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
```

```python
df.index.year
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(df.index.year, label='Fechas de contratacion')
plt.show()
```
